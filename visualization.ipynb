```
import pyspark.sql.functions as F
import pyspark.sql.types as T
def deleteNull_castToInt(x):
  
  if x == None:
    return 0
  if x == '(not set)':
    return 0
  if x == 'No':
    return 0
  else:
    return int(x)
udf_deleteNull_castToInt = F.udf(deleteNull_castToInt, T.IntegerType())
DF2 = DF.select(udf_deleteNull_castToInt("totals_timeOnSite").alias("time"), "date")
```

```
from pyspark.sql.functions import col, unix_timestamp, to_date
DF_date = DF.select((col('date')).cast("string").alias("date_cast"), udf_deleteNull_castToInt("totals_transactionRevenue").alias("revenue"))
DF_date.show()
```

```

DF_date = DF_date.withColumn('date_cast', to_date('date_cast', 'yyyyMMdd'))
DF_date.show()
```

```
from pyspark.sql.functions import month, dayofweek
DF_date = DF_date.withColumn("month", month("date_cast"))
DF_date = DF_date.withColumn("weekday", dayofweek("date_cast"))
DF_date.show()
```
