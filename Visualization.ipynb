```
import pyspark.sql.functions as F
import pyspark.sql.types as T
def deleteNull_castToFloat(x):
  
  if x == None:
    return 0.0
  if x == '(not set)':
    return 0
  if x == 'No':
    return 0.0
  else:
    y = len(x)
    if y>6:
      i = int(x[ : y-6])
      f = int(x[y-6:y])
    else:
      i = 0
      f = int(x)
    a = float(i)
    b = float(f)/1000000
    return a+b
udf_deleteNull_castToFloat = F.udf(deleteNull_castToFloat, T.FloatType())

def deleteNull_castToInt(x):
  
  if x == None:
    return 0
  if x == '(not set)':
    return 0
  if x == 'No':
    return 0
  else:
    return int(x)
udf_deleteNull_castToInt = F.udf(deleteNull_castToInt, T.IntegerType())

def deleteNull_castToInt_visitNumber(x):
  
  if x == None:
    return 0
  if x == '(not set)':
    return 0
  if x == 'No':
    return 0
  if len(x)>10:
    return 0
  else:
    return int(x)
udf_deleteNull_castToInt_visitNumber = F.udf(deleteNull_castToInt_visitNumber, T.IntegerType())
```

```
# Date and time features:

from pyspark.sql.functions import col, unix_timestamp, to_date
from pyspark.sql.functions import from_unixtime
from pyspark.sql.functions import month, dayofweek, hour, year, dayofyear, weekofyear
# Convert String to Date objects in "date" column, convert String to TimeStamp objects in "visitId" column

DF_date = DF.select((col('date')).cast("string").alias("date_cast"), udf_deleteNull_castToFloat("totals_transactionRevenue").alias("revenue"), (col('visitStartTime')).cast("string").alias("startTime_cast"), udf_deleteNull_castToInt("hits_hour").alias("hits_hour"))
DF_date = DF_date.withColumn('date_cast', to_date('date_cast', 'yyyyMMdd'))
# Extract further information from Date objects
DF_date = DF_date.withColumn("month", month("date_cast"))
DF_date = DF_date.withColumn("weekday", dayofweek("date_cast"))
DF_date = DF_date.withColumn("year", year("date_cast"))
DF_date = DF_date.withColumn("startTime", from_unixtime("startTime_cast"))
DF_date = DF_date.withColumn("hour", hour("startTime"))
DF_date = DF_date.withColumn("yearday", dayofyear("date_cast"))
DF_date = DF_date.withColumn("yearweek", weekofyear("date_cast"))
```

```
# Average revenue per day across year
display(DF_date.filter("revenue>0").groupBy("yearday").avg("revenue").orderBy("yearday"))
```
